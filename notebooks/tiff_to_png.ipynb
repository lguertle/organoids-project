```python
from PIL import Image
import os
from importlib import reload
import utility
reload(utility)
from utility import get_hard_disk_path

# Define the directories
initial_path = get_hard_disk_path("Laure")
print(initial_path)
directories = ["empty", "j2", "j8"]

# Create a function to convert TIFF to PNG
def convert_tiff_to_png(directory, initial_path):
    # Check if the directory exists
    directory = os.path.join(initial_path, directory)
    if not os.path.exists(directory):
        print(f"Directory '{directory}' not found.")
        return

    # Make a new directory to store the PNG files
    png_directory = f"{directory}_png"
    os.makedirs(png_directory, exist_ok=True)

    # Convert each TIFF file in the directory to PNG
    for filename in os.listdir(directory):
        if filename.endswith(".tiff"):
            with Image.open(os.path.join(directory, filename)) as img:
                # Change the file extension to .png
                png_filename = os.path.splitext(filename)[0] + ".png"
                # Save the image in PNG format
                img.save(os.path.join(png_directory, png_filename))
            print(f"Converted {filename} to PNG format.")

# Convert the images in each directory
for dir in directories:
    convert_tiff_to_png(dir, initial_path)
```
