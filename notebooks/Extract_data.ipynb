```python
def center_crop(img, dim):
    """
    Returns center cropped image
    Args:
    img: image to be center cropped
    dim: dimensions (width, height) to be cropped
    """
    width, height = img.shape[1], img.shape[0]

# process crop width and height for max available dimension

    mid_x, mid_y = int(width/2), int(height/2)
    cw2, ch2 = int(dim/2), int(dim/2)
    crop_img = img[mid_y-ch2:mid_y+ch2, mid_x-cw2:mid_x+cw2]
    return crop_img
```

```python
import os
import cv2

path = r"E:\reformed_data"
path_seg = r"E:\data_for_seg"

destination_paths = {
    'keep0': os.path.join(path_seg, 'keep0'),
    'keep1': os.path.join(path_seg, 'keep1'),
    'keep2': os.path.join(path_seg, 'keep2'),
    'receed0': os.path.join(path_seg, 'receed0'),
    'receed1': os.path.join(path_seg, 'receed1'),
    'split': os.path.join(path_seg, 'split'),
    'dead': os.path.join(path_seg, 'dead'),
    'empty': os.path.join(path_seg, 'empty')
}

# Walk through the directory and its subdirectories
for root, dirs, files in os.walk(path):
    for file in files:
        print(file)
        # You can do something with each file here
        # For example, print the file path
        file_path = os.path.join(root, file)
        if file_path.endswith(".tif"):
            folder_name = os.path.basename(root)
            file_name = os.path.splitext(os.path.basename(file_path))[0]
            print(file_path, folder_name)
            # Read the image
            image = cv2.imread(file_path, cv2.IMREAD_UNCHANGED)
            if image is not None:
                image = center_crop(image, 7000)
                image = cv2.resize(image, (1000,1000))
                print(os.path.join(destination_paths[folder_name], file_name + '.jpg'))
                cv2.imwrite(os.path.join(destination_paths[folder_name], file_name + '.jpg'), image)
            else:
                print(f"Failed to read image: {file_path}")
```
